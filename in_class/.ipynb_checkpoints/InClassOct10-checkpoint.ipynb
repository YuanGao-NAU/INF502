```python
class Actor():
    def __init__(self, name, date_of_birth, nationality):
        self.name = name
        self.date_of_birth = date_of_birth
        self.nationality = nationality

    def change_name(self, name):
        self.name = name

    def change_date_of_birth(self, date_of_birth):
        self.date_of_birth = date_of_birth

    def change_nationality(self, nationality):
        self.nationality = nationality

    def list_all_works(self, movies):
        name = self.name
        total = []
        for movie in movies:
            for actor in movie.cast:
                if name == actor.name:
                    total.append(movie.title)
        if(total == []):
            print("the actor %s didn't participante all the movies" %name)
        else:
            print(total)

    def __str__(self):
        return "name:" + self.name + "date_of_birth:" + self.date_of_birth + "nationality:" + self.nationality
```

```python
class Movie(Actor):
    def __init__(self, title, genres, year, my_review, cast, counter, borrowed, borrower_name):
        self.cast = cast
        self.title = title
        self.genres = genres
        self.year = year
        self.my_review = my_review
        self.counter = counter
        self.borrowed = borrowed
        if(borrowed == False):
            self.borrower = ''
        else:
            self.borrower = borrower_name

    def watching_the_movie(self):
        self.counter  = self.counter + 1

    def writing_a_review(self, my_review):
        self.my_review = my_review

    def change_genres(self, genres):
        self.genres = genres

    def change_year(self, year):
        self.year = year

    def change_my_review(self, my_review):
        self.my_review = my_review
    
    def change_cast(self, cast):
        self.cast = cast

    def borrowing_the_movie(self, borrower):
        self.borrowed = True
        self.borrower = borrower

    def returning_the_movie(self):
        self.borrowed = False
        self.borrower = ''

    def show_infos(self):
        str1 = ""
        for actor in self.cast:
            str1 = str1 + actor.name + ', '
        str1 = str1[0:-2]
        print(" title: %s\n \
genres: %s\n \
year: %s\n \
my_review: %s\n \
actors: %s\n \
counter: %d\n \
borrower: %s\n \
borrower: %s" %(self.title, str(self.genres), self.year, self.my_review, str1, self.counter, self.borrowed, self.borrower))

    def __str__(self):
        str1 = ''
        str2 = '    '
        for genre in self.genres:
            str1 = str1 + genre + ','
        str1 = str1[0:-1]
        for actor in self.cast:
            str2 = str2 + actor.name + "\n" + "    "
        return "Movie: " + self.title + "    " + "Year: " + str(self.year) + "\n" + \
            "Genre: " + str1 + '\n' + \
            "list of Actors: \n" + \
            str2
```
