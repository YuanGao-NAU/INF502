The class listed below is the class of the file ```ant_bait.csv```, and the items in the ```__init__``` function is the same as the items in the file. ```__str__``` function is used to return the formatted string required in the \#1. The ```replace_attribute``` is used to replace the ```species``` of ant_bait object with ```species``` from ant_species.

````python
class ant_bait:
    def __init__(self, month, year, plot, stake, species, abundance):
        self.month = month
        self.year = year
        self.plot = plot
        self.stake = stake
        self.species = species
        self.abundance = abundance

    def __str__(self):
        return self.species + " - " + self.month + ", " + self.year + ": " + self.abundance

    def replace_attribute(self, species):
        self.species = species
````

The class listed below is the class of the file ```ant_species.csv```, and the items in the ```__init__``` is the same as the items in the file. ```__str__``` is used to return the formatted string required in the \#1.

````python
class ant_species:
    def __init__(self, species_code, genus, alt_genus, species, alt_species, tribe, sub_family, Dissues):
        self.species_code = species_code
        self.genus = genus
        self.alt_genus = alt_genus
        self.species = species
        self.alt_species = alt_species
        self.tribe = tribe
        self.sub_family = sub_family
        self.Dissues = Dissues
    
    def __str__(self):
        return self.genus + " " + self.species
````

The two functions listed below are used to read the csv files. ```read_ant_bait``` is used to read ```ant_bait.csv```, and ```read_ant_species``` is used to read ```ant_species.csv```. In the two functions, a built in library ```csv``` is used. So, ```import csv``` should be added in the beginning of the file. 

````python
import csv
def read_ant_bait(filename):
    ant_bait_list = []
    fd = open(filename, 'r')
    reader = csv.reader(fd)
    for line in reader:
        ant_bait1 = ant_bait(line[0], line[1], line[2], line[3], line[4], line[5])
        ant_bait_list.append(ant_bait1)
    #print(ant_bait_list)
    ant_bait_list = ant_bait_list[1:]
    return ant_bait_list

def read_ant_species(filename):
    ant_species_list = []
    fd = open(filename, 'r')
    reader = csv.reader(fd)
    for line in reader:
        ant_species1 = ant_species(line[0], line[1], line[2], line[3], line[4], line[5], line[6], line[7])
        #print(ant_species1)
        ant_species_list.append(ant_species1)
    ant_species_list = ant_species_list[1:]
    return ant_species_list
````

This function is used to deal with the ```ant_bait_list``` and ```ant_species_list``` gained by the function ```read_ant_bait``` and ```read_ant_species```, respectively. I used two for loops to get match the ```species``` from ```ant_bait_list``` and ```species_code``` from ```ant_species_list```, and used the ```replace_attribute``` function of the class ```ant_bait``` to replace the ```species``` of ```ant_bait``` with  corresponding ```species``` from ```ant_species```.

````python
def replace_species(ant_bait_list, ant_species_list):
    for item1 in ant_bait_list:
        for item2 in ant_species_list:
            if item1.species == item2.species_code:
                item1.replace_attribute(item2.species)
````

The function below is the ```main``` function. I tested the methods of the two classes.

````python
if __name__ == "__main__":
    list0 = read_ant_bait("./ant_bait.csv")
    list1 = read_ant_species("./ant_species.csv")
    for item in list1:
        print(item)
    print("------------------------------------------------------------------------------")
    for item in list0:
        print(item)
    replace_species(list0, list1)
    print("------------------------------------------------------------------------------")
    for item in list0:
        print(item)
````
